```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
from statsmodels.graphics.mosaicplot import mosaic

# -----------------------
# 1. Load and Prepare Data
# -----------------------
# Replace with your actual file path if necessary
data = pd.read_excel('/home/aricept094/mydata/GhaderiDataSPSS(310Covid patient).xlsx')

# Map codes to readable labels
data['Hyperglycemia_label'] = data['Hyperglycemia'].map({0: 'Normoglycemia', 1: 'Hyperglycemia'})
data['Severity'] = data['covid_stage_binary'].map({0: 'Mild/Moderate', 1: 'Severe/Critical'})
data['Outcome_label'] = data['Outcome'].map({1: 'Survival', 2: 'Death'})

# -----------------------
# 2. Stacked Bar Chart: COVID-19 Severity by Glycemic Status
# -----------------------
# Prepare data for stacked bar chart
severity_counts = data.groupby(['Hyperglycemia_label', 'Severity']).size().reset_index(name='Count')
severity_pivot = severity_counts.pivot(index='Hyperglycemia_label', columns='Severity', values='Count')

# Plot Stacked Bar Chart
plt.figure(figsize=(8,6))
severity_pivot.plot(kind='bar', stacked=True, color=['#4c72b0', '#dd8452'])
plt.title('COVID-19 Severity Distribution by Glycemic Status')
plt.xlabel('Glycemic Status')
plt.ylabel('Number of Patients')
plt.legend(title='COVID-19 Severity')
plt.tight_layout()
plt.show()

# -----------------------
# 3. Grouped Bar Chart: Mortality (Outcome) by Glycemic Status
# -----------------------
# Prepare data for grouped bar chart
outcome_counts = data.groupby(['Hyperglycemia_label', 'Outcome_label']).size().reset_index(name='Count')

plt.figure(figsize=(8,6))
sns.barplot(x='Hyperglycemia_label', y='Count', hue='Outcome_label', data=outcome_counts, palette='Set2')
plt.title('COVID-19 Outcome by Glycemic Status')
plt.xlabel('Glycemic Status')
plt.ylabel('Number of Patients')
plt.legend(title='Outcome')
plt.tight_layout()
plt.show()

# -----------------------
# 4. Sankey Diagram: Flow from Glycemic Status to Severity to Outcome
# -----------------------
# Define node labels and mapping
source_labels = ['Normoglycemia', 'Hyperglycemia']
mid_labels = ['Mild/Moderate', 'Severe/Critical']
end_labels = ['Survival', 'Death']
all_nodes = source_labels + mid_labels + end_labels
node_indices = {node: i for i, node in enumerate(all_nodes)}

# Aggregate flows: From Glycemic Status to Severity
flow1 = data.groupby(['Hyperglycemia_label', 'Severity']).size().reset_index(name='Count')
# Aggregate flows: From Severity to Outcome
flow2 = data.groupby(['Severity', 'Outcome_label']).size().reset_index(name='Count')

# Build the Sankey diagram
fig = go.Figure(data=[go.Sankey(
    node = dict(
        pad = 15,
        thickness = 20,
        line = dict(color = "black", width = 0.5),
        label = all_nodes,
        color = ["#8ecae6", "#219ebc", "#ffb703", "#fb8500", "#a8dadc", "#e63946"]
    ),
    link = dict(
        source = (
            [node_indices[row['Hyperglycemia_label']] for _, row in flow1.iterrows()] +
            [node_indices[row['Severity']] for _, row in flow2.iterrows()]
        ),
        target = (
            [node_indices[row['Severity']] for _, row in flow1.iterrows()] +
            [node_indices[row['Outcome_label']] for _, row in flow2.iterrows()]
        ),
        value = list(flow1['Count']) + list(flow2['Count'])
    ))])

fig.update_layout(title_text="Patient Flow: Glycemic Status → COVID-19 Severity → Outcome", font_size=10)
fig.show()

# -----------------------
# 5. Mosaic Plot: Association Between Hyperglycemia, Severity, and Outcome
# -----------------------
# Create a contingency table with two dimensions (Hyperglycemia & Severity vs. Outcome)
contingency = pd.crosstab([data['Hyperglycemia_label'], data['Severity']], data['Outcome_label'])

plt.figure(figsize=(10,8))
mosaic(contingency.stack(), title='Mosaic Plot: Hyperglycemia, COVID-19 Severity, and Outcome')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import plotly.graph_objects as go

# 1. Load and Prepare Data
file_path = '/home/aricept094/mydata/GhaderiDataSPSS(310Covid patient).xlsx' # Replace with your actual file path
df = pd.read_excel(file_path)

# Recode variables for better labels
df['Hyperglycemia_Label'] = df['Hyperglycemia'].map({0: 'Normoglycemia', 1: 'Hyperglycemia'})
df['Covid_Stage_Label'] = df['covid_stage_binary'].map({0: 'Mild/Moderate', 1: 'Severe/Critical'})
df['Outcome_Label'] = df['Outcome'].map({1: 'Survival', 2: 'Death'})

# 2. Calculate Counts for Sankey Diagram Links

# Glycemia to Covid Stage
glycemia_to_stage_counts = df.groupby(['Hyperglycemia_Label', 'Covid_Stage_Label']).size().reset_index(name='counts')

# Covid Stage to Outcome
stage_to_outcome_counts = df.groupby(['Covid_Stage_Label', 'Outcome_Label']).size().reset_index(name='counts')

# Total counts for each Glycemia group (for proportions if needed - not used in this basic Sankey but could be added)
glycemia_counts = df['Hyperglycemia_Label'].value_counts()


# 3. Calculate Odds Ratios (and prepare text for annotations)

def calculate_odds_ratio(group1_name, group2_name, outcome_variable, favorable_outcome_cat, unfavorable_outcome_cat, df):
    """Calculates Odds Ratio and 95% CI for binary outcomes."""
    table = pd.crosstab(df[group1_name] == group2_name, df[outcome_variable])

    a = table.loc[True, unfavorable_outcome_cat] # Unfavorable outcome in group of interest (e.g., Hyperglycemia & Severe)
    b = table.loc[True, favorable_outcome_cat]   # Favorable outcome in group of interest (e.g., Hyperglycemia & Mild/Moderate)
    c = table.loc[False, unfavorable_outcome_cat] # Unfavorable outcome in reference group (e.g., Normoglycemia & Severe)
    d = table.loc[False, favorable_outcome_cat]   # Favorable outcome in reference group (e.g., Normoglycemia & Mild/Moderate)

    odds_ratio = (a * d) / (b * c)
    return odds_ratio

# Odds Ratio for Severe/Critical COVID-19 in Hyperglycemia vs. Normoglycemia
or_severe_covid = calculate_odds_ratio(
    'Hyperglycemia_Label', 'Hyperglycemia', 'Covid_Stage_Label', 'Mild/Moderate', 'Severe/Critical', df
)
or_severe_covid_text = f"OR={or_severe_covid:.2f}"

# Odds Ratio for Mortality in Hyperglycemia vs. Normoglycemia
or_mortality = calculate_odds_ratio(
    'Hyperglycemia_Label', 'Hyperglycemia', 'Outcome_Label', 'Survival', 'Death', df
)
or_mortality_text = f"OR={or_mortality:.2f}"


# 4. Define Sankey Diagram Nodes and Links

# Node labels - ordered for visual flow
node_labels = ["Normoglycemia", "Hyperglycemia", "Mild/Moderate", "Severe/Critical", "Survival", "Death"]

# Node colors (optional, for visual distinction)
node_colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b'] # Example colors


# Link sources, targets, and values - based on counts

# Glycemia to Stage links
link_sources_glycemia_stage = []
link_targets_glycemia_stage = []
link_values_glycemia_stage = []
link_label_glycemia_stage = [] # Use label instead of text


source_map_glycemia = {"Normoglycemia": 0, "Hyperglycemia": 1}
target_map_stage = {"Mild/Moderate": 2, "Severe/Critical": 3}


for index, row in glycemia_to_stage_counts.iterrows():
    link_sources_glycemia_stage.append(source_map_glycemia[row['Hyperglycemia_Label']])
    link_targets_glycemia_stage.append(target_map_stage[row['Covid_Stage_Label']])
    link_values_glycemia_stage.append(row['counts'])
    link_label_glycemia_stage.append(str(row['counts'])) # Count as label


# Stage to Outcome links
link_sources_stage_outcome = []
link_targets_stage_outcome = []
link_values_stage_outcome = []
link_label_stage_outcome = [] # Use label instead of text


source_map_stage_outcome = {"Mild/Moderate": 2, "Severe/Critical": 3}
target_map_outcome = {"Survival": 4, "Death": 5}


for index, row in stage_to_outcome_counts.iterrows():
    link_sources_stage_outcome.append(source_map_stage_outcome[row['Covid_Stage_Label']])
    link_targets_stage_outcome.append(target_map_outcome[row['Outcome_Label']])
    link_values_stage_outcome.append(row['counts'])
    link_label_stage_outcome.append(str(row['counts'])) # Count as label


# Combine all links
link_sources = link_sources_glycemia_stage + link_sources_stage_outcome
link_targets = link_targets_glycemia_stage + link_targets_stage_outcome
link_values = link_values_glycemia_stage + link_values_stage_outcome
link_label = link_label_glycemia_stage + link_label_stage_outcome # Use label here

# 5. Create Sankey Diagram with Plotly

fig = go.Figure(data=[go.Sankey(
    node = dict(
      pad = 15,
      thickness = 20,
      line = dict(color = "black", width = 0.5),
      label = node_labels,
      color = node_colors
    ),
    link = dict(
      source = link_sources,
      target = link_targets,
      value = link_values,
      label = link_label, # Use label instead of text
      hoverinfo = 'skip', # Disable hover info for cleaner static image
      line=dict(color="black", width=0.1) # Optional link lines for better definition
  ))])


# 6. Add Annotations for Odds Ratios (Positioning might need adjustment)

fig.add_annotation(
    x=0.3, y=0.5,  # Adjust x and y for horizontal position between Glycemia and Stage
    xref="paper", yref="paper",
    text=or_severe_covid_text,
    showarrow=False,
    font=dict(size=12, color="black")
)

fig.add_annotation(
    x=0.7, y=0.5, # Adjust x and y for horizontal position between Stage and Outcome
    xref="paper", yref="paper",
    text=or_mortality_text,
    showarrow=False,
    font=dict(size=12, color="black")
)


# 7. Update Layout for better presentation

fig.update_layout(
    title_text="<b>Flow of COVID-19 Patients by Glycemic Status, Severity, and Outcome</b><br><sup>Odds Ratios are shown for Severe/Critical COVID-19 and Mortality in Hyperglycemic patients</sup>",
    font_size=12,
    title_font_size=14,
    margin=dict(l=50, r=50, b=80, t=100), # Adjust margins for title and annotations
)


fig.show()

# To save as a static image for research paper (e.g., png, svg)
# import plotly.io as pio
# pio.write_image(fig, "covid_sankey_with_or.png", format="png", width=1200, height=800)
```

```python
import pandas as pd
import plotly.graph_objects as go

# Load the data
file_path = '/home/aricept094/mydata/GhaderiDataSPSS(310Covid patient).xlsx'
df = pd.read_excel(file_path)

# Rename columns for easier use (optional but recommended)
df.rename(columns={
    'Hyperglycemia': 'Hyperglycemia',
    'covid_stage_binary': 'COVID_Stage',
    'Outcome': 'Outcome'
}, inplace=True)

# Recode categorical variables to descriptive labels for the diagram
df['Hyperglycemia_Label'] = df['Hyperglycemia'].map({0: 'Normoglycemia', 1: 'Hyperglycemia'})
df['COVID_Stage_Label'] = df['COVID_Stage'].map({0: 'Mild/Moderate', 1: 'Severe/Critical'})
df['Outcome_Label'] = df['Outcome'].map({1: 'Survival', 2: 'Death'})

# Calculate value counts for each category (not used in node labels anymore)
hyper_counts = df['Hyperglycemia_Label'].value_counts()
stage_counts = df['COVID_Stage_Label'].value_counts()
outcome_counts = df['Outcome_Label'].value_counts()

# Define nodes - Numbers removed from labels (common for all plots)
base_nodes = [
    {'label': "Normoglycemia", 'category': 'Normoglycemia'},
    {'label': "Hyperglycemia", 'category': 'Hyperglycemia'},
    {'label': "Mild/Moderate", 'category': 'Mild/Moderate'},
    {'label': "Severe/Critical", 'category': 'Severe/Critical'},
    {'label': "Survival", 'category': 'Survival'},
    {'label': "Death", 'category': 'Death'}
]

# Define links - count transitions (common for all plots)
ng_mild_moderate = len(df[(df['Hyperglycemia_Label'] == 'Normoglycemia') & (df['COVID_Stage_Label'] == 'Mild/Moderate')])
ng_severe_critical = len(df[(df['Hyperglycemia_Label'] == 'Normoglycemia') & (df['COVID_Stage_Label'] == 'Severe/Critical')])
hg_mild_moderate = len(df[(df['Hyperglycemia_Label'] == 'Hyperglycemia') & (df['COVID_Stage_Label'] == 'Mild/Moderate')])
hg_severe_critical = len(df[(df['Hyperglycemia_Label'] == 'Hyperglycemia') & (df['COVID_Stage_Label'] == 'Severe/Critical')])

mild_moderate_survival = len(df[(df['COVID_Stage_Label'] == 'Mild/Moderate') & (df['Outcome_Label'] == 'Survival')])
mild_moderate_death = len(df[(df['COVID_Stage_Label'] == 'Mild/Moderate') & (df['Outcome_Label'] == 'Death')])
severe_critical_survival = len(df[(df['COVID_Stage_Label'] == 'Severe/Critical') & (df['Outcome_Label'] == 'Survival')])
severe_critical_death = len(df[(df['COVID_Stage_Label'] == 'Severe/Critical') & (df['Outcome_Label'] == 'Death')])

base_links = [
    {'source': 0, 'target': 2, 'value': ng_mild_moderate, 'label': ""},
    {'source': 0, 'target': 3, 'value': ng_severe_critical, 'label': ""},
    {'source': 1, 'target': 2, 'value': hg_mild_moderate, 'label': ""},
    {'source': 1, 'target': 3, 'value': hg_severe_critical, 'label': ""},
    {'source': 2, 'target': 4, 'value': mild_moderate_survival, 'label': ""},
    {'source': 2, 'target': 5, 'value': mild_moderate_death, 'label': ""},
    {'source': 3, 'target': 4, 'value': severe_critical_survival, 'label': ""},
    {'source': 3, 'target': 5, 'value': severe_critical_death, 'label': ""}
]

# Define color palette for nodes (consistent Survival color)
node_color_palette = {
    "Normoglycemia": '#7CA1CC',  # Sky Blue
    "Hyperglycemia": '#D55E00',   # Orange
    "Mild/Moderate": '#F0B98D',  # Peach
    "Severe/Critical": '#2E5984', # Navy
    "Survival": '#A9DFBF',      # Pastel Green (Survival color - fixed)
    "Death": '#000000'         # Black
}


# Define multiple link colors to test
link_colors = {
    "Links_Grey": 'rgba(128, 128, 128, 0.5)',       # Original Grey
    "Links_VeryLightGrey": 'rgba(200, 200, 200, 0.5)', # Very Light Grey
    "Links_NatureLightGrey": '#F7F7F7',             # Nature Palette Light Grey (Background Neutral)
    "Links_Peach": '#F0B98D',                        # Peach (from Nature Palette - Secondary Accent - Experiment)
    "Links_SkyBlue": '#7CA1CC',                         # Sky Blue (from Nature Palette - Secondary Accent - Experiment)
    "Links_VeryPaleBlue": '#E0FFFF'                  # Very Pale Blue (New - for much lighter blue links)

    # Add more link colors here if you want to try others
}


# Function to create and save Sankey diagram with a given link color
def create_sankey_diagram_link_color(link_color_name, link_color_code):
    fig = go.Figure(data=[go.Sankey(
        node = dict(
          pad = 15,
          thickness = 20,
          line = dict(color = "black", width = 0.5),
          label = [node['label'] for node in base_nodes],
          color = [node_color_palette[node['category']] for node in base_nodes] # Consistent node colors
        ),
        link = dict(
          source = [link['source'] for link in base_links],
          target = [link['target'] for link in base_links],
          value = [link['value'] for link in base_links],
          label = [link['label'] for link in base_links],
          color =  [link_color_code] * len(base_links), # Apply link color here
          line = dict(color = "black", width = 0.1),
          hovertemplate='Source: %{source.label}<br>Target: %{target.label}<br>Value: %{value}<extra></extra>'
      ))]
    )

    fig.update_layout(
        title_text=f"COVID-19 Patient Outcomes - Link Color: {link_color_name}",
        font_size=12,
        showlegend=False,
    )

    # Save the figure with 900 DPI
    filename = f"sankey_diagram_Links_{link_color_name}_900dpi.png" # Updated filename to 900dpi
    fig.write_image(filename, scale=9) # scale=9 for approximately 900 DPI
    print(f"Sankey diagram saved as {filename} with approximately 900 DPI and Link Color: {link_color_name}")
    return filename # Return filename for easy reference if needed


# Loop through each link color and generate a plot
saved_files_links = {}
for link_color_name, link_color_code in link_colors.items():
    filename = create_sankey_diagram_link_color(link_color_name, link_color_code)
    saved_files_links[link_color_name] = filename


print("\nAll Sankey diagrams generated and saved (Link Color Variations) at 900 DPI:")
for name, file in saved_files_links.items():
    print(f"- {name}: {file}")
```
